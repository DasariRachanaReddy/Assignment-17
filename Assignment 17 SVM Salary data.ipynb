```python
# SVM Classification
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report


from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_val_score

```

```python
Train = pd.read_csv('SalaryData_Train(1).csv')
```

```python
Test = pd.read_csv('SalaryData_Test(1).csv')
```

```python
# Convert categorical variables into numerical values
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
categorical_columns = ["workclass", "education", "maritalstatus", "occupation", "relationship", "race", "sex", "native"]
for column in categorical_columns:
    Train[column] = encoder.fit_transform(Train[column])
```

```python
encoder = LabelEncoder()
categorical_columns = ["workclass", "education", "maritalstatus", "occupation", "relationship", "race", "sex", "native"]
for column in categorical_columns:
    Test[column] = encoder.fit_transform(Test[column])
```

```python
##Capturing the column names which can help in futher process
colnames = Train.columns
colnames
```

```python
x_train = Train[colnames[0:13]]
y_train = Train[colnames[13]]
x_test = Test[colnames[0:13]]
y_test = Test[colnames[13]]
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
# Train the model
clf = SVC()
clf.fit(x_train, y_train)
```

```python
# Evaluate the model
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

y_pred = clf.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
